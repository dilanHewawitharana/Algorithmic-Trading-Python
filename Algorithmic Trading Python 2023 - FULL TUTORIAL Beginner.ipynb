```python
# copy the AAPL values from close and convert into data frame
aapl = close.AAPL.copy().to_frame()
aapl
```

```python
# Shift aapl column by 1
#aapl.shift(periods=1)
```

```python
# shift aapl column by 1 and store into lag1 column
aapl["lag1"]=aapl.shift(periods=1)
aapl
```

```python
# subtract AAPL value by lag1 value
aapl["Diff"]=aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
aapl["% Change"]=aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
aapl
```

```python
# Diff2 shows the difference happens by one day
aapl["Diff2"]=aapl.AAPL.diff(periods=1)
aapl
```

```python
# % change 2 shows the percentage diff happens by one day
aapl["% change 2"]=aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
# Delete column
del aapl["lag1"]
del aapl["Diff"]
del aapl["% Change"]
del aapl["Diff2"]
aapl
```

```python
# rename column
aapl.rename(columns = {'% change 2':'Change'}, inplace=True)
aapl
```

```python
# resample data with respect to month
aapl.AAPL.resample("M").last()
```

```python
# resample data with respect business day for of month
aapl.AAPL.resample("BM").last()
```

```python
# resample data with respect business day for of month, percentage change happens per each month
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

```python
del aapl["Change"]
aapl
```

```python
# calculate percentage chage and drop NAN values if exist
ret = aapl.pct_change().dropna()
ret
```

```python
ret.info()
```

```python
# plot histogram
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

```python
# calculate mean
daily_mean_ret = ret.mean()
daily_mean_ret
```

```python
# calculate variance
var_daily = ret.var()
var_daily
```

```python
# standard deviation
std_daily = np.sqrt(var_daily)
std_daily
```

```python
# Another way to caculate standard variation
std_daily2 = ret.std()
std_daily2
```

```python
# calculate annual mean return
annual_mean_ret=daily_mean_ret*252
annual_mean_ret
```

```python
# calculate annual var return
annual_var_ret=var_daily*252
annual_var_ret
```

```python
# calculate annual std return
annual_std_ret=np.sqrt(annual_var_ret)
annual_std_ret
```
