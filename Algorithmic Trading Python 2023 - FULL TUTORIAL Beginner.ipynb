```python
# Download SPY, APPLE, KO companies stocks data from ybinance from 2010 to 2021
ticker=["SPY", "AAPL", "KO"]
stocks = yf.download(ticker, start = "2010-01-01", end="2021-01-01")
```

```python
stocks
```

```python
# show sample data to get an idea about stocks
stocks.head()
```

```python
# show stock information
stocks.info()
```

```python
# save stocks into csv
stocks.to_csv("stocksYT.csv")
```
